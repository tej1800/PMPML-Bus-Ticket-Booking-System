```python
import datetime
import random

print("Welcome to PMPML Pune")
x = datetime.datetime.now()
print(x)

stop = ["MaNaPa(0)", "Pune station(1)", "Wadia college(2)", "Yervda(3)", "Viman nagar(4)", "IANI institute(5)"]

start = int(input("Enter Boarding station: "))
end = int(input("Enter End station: "))
ticket = (end - start) * 5
print("Your ticket is:", ticket)

payment = input("Enter your payment mode:(1.Cash/2.UPI/3.OTP): ")

if payment == "1":
    payment = "Cash"
elif payment == "2":
    payment = "UPI"
elif payment == "3":
    payment = "OTP"
else:
    print("Invalid option.. ")

if payment == "Cash":
    print(ticket, "Rupees collected")
    print("Happy journey!")
elif payment == "UPI":
    print("generating QR Code.......")
    side = 3
    for i in range(side):
        for j in range(side):
            if(i == 0 or i == side-1 or j == 0 or j == side-1):
                print('*', end=' ')
            else:
                print(' ', end=' ')
        print()
    list = ["****@axisbank", "****@sbibank", "****@hdfcbank"]
    random_item = random.choice(list)
    print(ticket, "Rupees received by", random_item)
    print("Happy journey")
elif payment == "OTP":
    Mob_no = int(input("Enter Mobile number: "))
    otp = random.randint(1000, 9999)
    print("OTP sent to", Mob_no, "is **", otp, "**")
    user_otp = int(input("Enter OTP received by user: "))
    if user_otp == otp:
        print("Payment successful....Enjoy your journey!")
    else:
        print("Invalid OTP..Payment failed."" \nPlease try again later")

if payment in ["Cash", "UPI", "OTP"]:
    print("\n------------------------------------------------------")
    print("|               TICKET INFORMATION                   |")
    print("------------------------------------------------------")
    print("|   Welcome to PMPML Pune                            |")
    print("|   Date & Time: ", x)
    print("|   Start Station: ", stop[start])
    print("|   End Station: ", stop[end])
    print("|   Ticket Price: ", ticket, "Rupees")
    print("|   Payment Mode: ", payment)
    print("|   Have a Safe and Happy Journey!                   |")
    print("------------------------------------------------------")
```
